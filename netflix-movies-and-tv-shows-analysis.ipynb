# Netflix Movies and TV Shows Analysis
Welcome to this insightful analysis of Netflix movies and TV shows. We'll dive deep into the data to uncover interesting patterns and trends. Let's get started!

```python
# Suppress warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot style
sns.set(style='whitegrid')
```

```python
# Load the dataset
file_path = '/kaggle/input/netflix-movies/netflix_full.csv'
netflix_df = pd.read_csv(file_path)
```

## Data Overview
Let's take a quick look at the structure of our dataset.

```python
# Display the first few rows of the dataset
netflix_df.head()
```

## Data Cleaning
Before we dive into the analysis, let's clean the data. We'll handle missing values and ensure data types are correct.

```python
# Check for missing values
netflix_df.isnull().sum()
```

```python
# Fill missing values in 'creator', 'genres', and 'country' with 'Unknown'
netflix_df['creator'].fillna('Unknown', inplace=True)
netflix_df['genres'].fillna('Unknown', inplace=True)
netflix_df['country'].fillna('Unknown', inplace=True)
```

```python
# Verify that there are no more missing values
netflix_df.isnull().sum()
```

## Exploratory Data Analysis (EDA)
Now that our data is clean, let's explore it! We'll start with some basic visualizations.

```python
# Distribution of content types (Movies vs TV Shows)
plt.figure(figsize=(8, 6))
sns.countplot(data=netflix_df, x='type', palette='viridis')
plt.title('Distribution of Content Types')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()
```

```python
# Distribution of ratings
plt.figure(figsize=(12, 6))
sns.countplot(data=netflix_df, x='rating', palette='viridis', order=netflix_df['rating'].value_counts().index)
plt.title('Distribution of Ratings')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```
